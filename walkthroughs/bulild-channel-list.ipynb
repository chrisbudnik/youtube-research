## **Finding channels with `YoutubeSearch`**
- to ensure we dont't cross the limit of api calls, I created a `sample` that consists of top search terms saved in `search-terms.txt`
- with `collect_best_ranking_channels()` method I first have collected top ranking videos and then extracted channel names and ids


```python
sample = search_terms[:5]
search = YouTubeSearch(keywords=sample)
```

```python
channels = search.collect_best_ranking_channels(max_results = 50, order_by="viewCount", only_unique=True)
```

```python
print(f"With youtube search we found {len(channels)} unique channels", "\n")

print("Printing top 10 results")
pprint(channels[:5])
```

## **Saving Channel Names**

```python
with open('channel_names.txt', 'w') as file:
    for channel in channels:
        file.write(f'{channel.channel_name}\n')
```
