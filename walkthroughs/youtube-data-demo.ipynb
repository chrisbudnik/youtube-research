## **`Video` Class**

*description* </br>
Provides simple interface for interaction with video data </br>
Key features:
- based on id, deteriming video name
- calculate video length with len() function
- access static video properties


```python
id = "bmWv2Kn12fM" # joshua mayo, 12 Websites To Make Easy Money Daily ($200/Day)
id2 = "gvNUAvBiSOw" # graham stephen, BREAKING: The Federal Reserve Pivot (Major Changes Explained)

video = yt.Video(id)

# check video representation
print(video, "\n")

# check video lenghh
print("Video length is equal to: ", len(video), " seconds" "\n")

# Check if videos are equal
video2 = yt.Video(id2)
print(video == video2, "\n")

# Check video channel id
print(video.channel_id, "\n")

# check get video response method (inherited from YouTubeAPI)
pprint(video.get_video_response(video.video_id, part="statistics"))
print("\n")

# check get video properties method 
print(video.get_video_properties(), "\n") 

# check get video stats method
print(video.get_video_stats(), "\n")

# check video transcript
print(video.get_video_transcript())
```

## **`Channel` Class**

## **`YoutubeSearch` Class**

*description* </br>
</br>
Key features:
- based on id, deteriming video name

```python
from datetime import datetime, timezone, timedelta

search = yt.YouTubeSearch(["how to make money online"])

# execute search - raising execptions when type is not valid
#print(search.execute_search("playlist"))

# execute search - video base params
print(search.execute_search("video"), "\n")

# execute search - all params modified
print(search.execute_search(type="video", max_results=3, published_after=(datetime.now(timezone.utc) - timedelta(days=30)).strftime("%Y-%m-%dT%H:%M:%SZ"), order_by='viewCount'), "\n")

# execute search - channels
print(search.execute_search("channel"), "\n")

# collect_exact_terms method
channels_to_check_ids = yt.YouTubeSearch(["Graham Stephan", "joshua mayo", "graham stephen wrong name"])
print(channels_to_check_ids.collect_exact_terms(), "\n")

# collect best ranking channels method
print(search.collect_best_ranking_channels(max_results=3))
```

## **`YoutubeDataCollector` Class**

*description* </br>
</br>
Key features:
- based on id, deteriming video name

```python
channel_ids = ["UCX6OQ3DkcsbYNE6H8uQQuVA", "UC-lHJZR3Gqxm24_Vd_AJ5Yw"]
video_ids = ["bmWv2Kn12fM", "gvNUAvBiSOw"]

collector = yt.YouTubeDataCollector(channel_ids=channel_ids, video_ids=video_ids)

# check data collection from chanels
data_from_channels = collector.collect_data_from_channels(max_videos=3)

# check data collection from videos
data_from_videos = collector.collect_data_from_videos()
```

```python
# print video ids from selected channels
print([item["video_id"] for item in data_from_channels], "\n")

# print video views from selected videos
print([item["views"] for item in data_from_videos], "\n")
```
